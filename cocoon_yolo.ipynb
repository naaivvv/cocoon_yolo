```python
import os

# Define the paths to your label folders
paths = [
    '/content/drive/MyDrive/cocoon_yolo/cocoons_dataset_1/train/labels',
    '/content/drive/MyDrive/cocoon_yolo/cocoons_dataset_1/valid/labels'
]

for path in paths:
    cache_file = os.path.join(path, 'labels.cache')
    if os.path.exists(cache_file):
        try:
            os.remove(cache_file)
            print(f"Successfully deleted: {cache_file}")
        except Exception as e:
            print(f"Failed to delete {cache_file}: {e}")
    else:
        print(f"No cache file found at: {cache_file}")
```

```python
from ultralytics import YOLO
import torch

MODEL_NAME = 'yolo11n.pt'
# Updated path to point to Google Drive
DATA_YAML = '/content/drive/MyDrive/cocoon_yolo/cocoons_dataset_1/data.yaml'
IMG_SIZE = 512

def train_custom_yolo():
    print(f"Loading {MODEL_NAME}...")
    model = YOLO(MODEL_NAME)

    model.train(
        data=DATA_YAML,
        epochs=200,
        patience = 20,
        imgsz=IMG_SIZE,
        batch=-1,
        cache=True,
        # Saves the training run directly to your Drive so you don't lose it when Colab disconnects
        project='/content/drive/MyDrive/runs/detect',
        name='cocoon_train',
        exist_ok=True,
        augment=True,
        cos_lr=True,
        save=True
    )

    print("Training Complete!")

if __name__ == '__main__':
    train_custom_yolo()
```
